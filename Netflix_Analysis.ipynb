```python
#3) How many Movies & TV Shows are in the dataset ? Show with Bar Graph.
```

```python
#method-1
type_count=df.type.value_counts()
sns.barplot(x=type_count.index, y=type_count.values, palette='pastel')
```

```python
#Method-2:using groupby

type_count=df.groupby('type').type.count()
sns.barplot(x=type_count.index, y=type_count.values, palette='muted')
```
